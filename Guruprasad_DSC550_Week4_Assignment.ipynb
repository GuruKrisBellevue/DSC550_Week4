### 1.Load the data as a Pandas data frame and ensure that it imported correctly

```python
# Importing the auto-mpg dataset from the local directory
path=os.getcwd()
auto_mpg_path=path+"\\auto-mpg.csv"
# creating a dataframe for the Auto MPG dataset
auto_mpg_orig_df=pd.read_csv(auto_mpg_path)
# Printing top 5 rows from the Auto MPG dataset
auto_mpg_orig_df.head()
```

### 2. Begin by prepping the data for modeling:

#### 2.1. Remove the car name column.

```python
# Taking a copy of the dataframe
auto_mpg_df1=auto_mpg_orig_df.copy()
```

```python
# Getting the list of all columns
auto_mpg_df1.columns
```

```python
# Dropping the column car name from the dataframe
auto_mpg_df1=auto_mpg_df1.drop(['car name'],axis=1)
```

```python
# Printing top 5 rows after car name column is dropped
auto_mpg_df1.head()
```

#### 2.2. The horsepower column values likely imported as a string data type. Figure out why and replace any strings with the column mean.

```python
# Running the info function to get the datatypes of all columns in the dataframe
auto_mpg_df1.info()
# The output indicates the horsepower is imported as String column.
```

```python
print(f"""Nulls in each column of the dataframe:  
{auto_mpg_df1.isnull().sum()}

Min Value:  {auto_mpg_df1["horsepower"].min()}

Max Value: {auto_mpg_df1["horsepower"].max()}
""")
```

#### Reason the horse power column is loaded as string is, because of the presence of non numeric values such as "?". This is being treated below.

```python
# Using to_numeric to convert all values of HorsePower to Number. The Non-numeric values wil not throw exception
auto_mpg_df1["horsepower"]=auto_mpg_df1["horsepower"].apply(pd.to_numeric, errors='coerce')
```

```python
# Getting the rows that had non numeric values
non_numeric_hp_rows1=auto_mpg_df1[auto_mpg_df1["horsepower"].isnull()]
non_numeric_hp_rows1
```

```python
# Getting the mean of the  horsepower 
hp_mean=auto_mpg_df1["horsepower"].mean()
print(f"The mean of HorsePower column is {hp_mean}")
# Usinf fillna function to replace all null values with mean of Horse Power
auto_mpg_df1["horsepower"]=auto_mpg_df1["horsepower"].fillna(hp_mean)
```

```python
# Validating the presence of Non-numeric values after replacing nulls with Mean value
non_numeric_hp_rows2=auto_mpg_df1[auto_mpg_df1["horsepower"].isnull()]
# Results indicate all Non numeric values are treated 
non_numeric_hp_rows2
```

```python
# Running the info function to get the datatypes of all columns in the dataframe
auto_mpg_df1.info()
```

#### 2.3. Create dummy variables for the origin column.

```python
# Using Pandas get_dummies method to create dummy variables on the Origin column and dropping
# the first dummy column to avoid multi-collinearity 
auto_mpg_df2=pd.get_dummies(auto_mpg_df1, columns=["origin"],drop_first=True)
auto_mpg_df2.head()
```

```python
# Using One hot encoding method (other way of doing this)
one_hot_encoder = LabelBinarizer()
# Creating dummies variable using the one hot encoder
origin_dummies2=one_hot_encoder.fit_transform(auto_mpg_df1["origin"])
# Naming the columns to be used in the dataframe
col_names2=["origin_1","origin_2","origin_3"]
dummies_df=pd.DataFrame(origin_dummies2,columns=col_names2)
dummies_df
```

### 3. Create a correlation coefficient matrix and/or visualization. Are there features highly correlated with mpg?

```python
# Creating correlation matrix of the dataframe and rounding the decimal to 2 digits
corr_matrix = auto_mpg_df2.corr().round(2)
corr_matrix
```

```python
# Creating a HeatMap using Seaborn with annotation turned on 
sns.heatmap(corr_matrix, annot=True, vmax=1, vmin=-1, center=0, cmap='vlag')
# setting the size of the plot
sns.set(rc={"figure.figsize": (10, 10)})
plt.show()
```

```python
# Creating a correlation mask to mask the redundant values and the values along the diagonal which are 1
corr_mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
# Creating the heatmap with the Mask applied for easy readability
sns.heatmap(corr_matrix, annot=True, vmax=1, vmin=-1, center=0, cmap='vlag', mask=corr_mask)
sns.set(rc={"figure.figsize": (10, 10)})
plt.show()
```

```python
# Unstacking the Matrix results and extracting the results for mpg variable
corr_matrix_mpg=corr_matrix.unstack()["mpg"]
# Filtering the variables that have correlation coefficient greater than 0.5(includes both postive and negative)
corr_matrix_mpg[abs(corr_matrix_mpg) >= 0.5]
```

#### ***Interpretation:*** From the results, the column mpg is positively correlated with the 'Model year' and origin. Results also indicate that it has strong negative correlation with the variables weight , displacement, cylinders and horsepower.

### 5. Randomly split the data into 80% training data and 20% test data, where your target is mpg.

```python
# Creating the variables for the Data and the Target to be used for the model
features=auto_mpg_df2.drop(['mpg'],axis=1).values
target=auto_mpg_df2['mpg'].values
```

```python
# Importing from scikit learn
from sklearn.model_selection import train_test_split
# Splitting the data and the target based on 80 , 20 split using the train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(features,target,test_size=0.2) 
```

```python
# Printing the size of the Train and Test sets
print(f""" 
The shape of Features: {features.shape}
The shape of Target: {target.shape}
The shape of X_Train: {X_train.shape}
The shape of X_Test: {X_test.shape}
""")
```

### 6. Train an ordinary linear regression on the training data.

```python
# Importing Linear Regression from Scikit learn
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
# Creating a pipeline by first standardizing the data and then creating Regression model
pipe_lr=Pipeline(
[
    ('scalar',StandardScaler()),
    ('reg',LinearRegression())
])
# Training the data by fitting the model on X_train and Y_train 
pipe_lr.fit(X_train, Y_train)
```

### 7.Calculate R2, RMSE, and MAE on both the training and test sets and interpret your results.

```python
# Calculating the metrics using a function
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
def create_regression_metrics(y_obj,predicted_tgt):
    """
    This function takes the Predicted values as input and returns the results of various metrics like R2, 
    MAE, MSE and RMSE
    """
    # using r2_score to calculate the R2
    r2=r2_score(y_obj, predicted_tgt)
    print(f"R2: {r2} ")
    # using mean_absolute_error functionto calculate the MAE
    mae=mean_absolute_error(y_obj,predicted_tgt)
    print(f"MAE: {mae}")
    # using mean_squared_error function to calculate the MSE
    mse=mean_squared_error(y_obj,predicted_tgt)
    print(f"MSE: {mse}")
    # using mean_squared_error function to calculate the RMSE by taking square root of MSE
    rmse= np.sqrt(mean_squared_error(y_obj,predicted_tgt))
    print(f"RMSE: {rmse}")
    return (r2,mae,mse,rmse)
```

```python
# Creating a variable for Predicted value for Linear Regression Training set
pred_train_olr=pipe_lr.predict(X_train)
# Getting the Metrics for Ordinary Linear Regression Training set
r2_train_olr,mae_train_olr,mse_train_olr,rmse_train_olr=create_regression_metrics(Y_train,pred_train_olr)
```

```python
# Creating a variable for Predicted value for Linear Regression Test set
pred_test_olr=pipe_lr.predict(X_test)
# Getting the Metrics for Ordinary Linear Regression Test set
r2_test_olr,mae_test_olr,mse_test_olr,rmse_test_olr=create_regression_metrics(Y_test,pred_test_olr)
```

#### ***Interpretation of the results:*** The R2 represents the goodness of the fit of the model, how good the regression line fits for the dataset. It is also called Coefficient of determination and it ranges between 0 and 1. In this case the value of 0.81 represents that the 81% of the variablity of the target variable is explained by the regression model.

#### The Mean squared error is the squared value of the difference between the actual value and the predicted values by the model. Taking its square root, we get RMSE. It can range between 0 and infinity.  Smaller RMSE values indicates a better model, hence in this case the RMSE of 3.64 suggests the model is better

#### Mean Absolute Error is the sum of absolute difference between the actual values and the values predicted by the model. It can range between 0 and infinity.  Smaller values indicates better model. In this model, the value of 2.62 suggests the model can be a better fit.

### 8. Pick another regression model and repeat the previous two steps. Note: Do NOT choose logistic regression as it is more like a classification model.

```python
# Using Lasso model to train the dataset
from sklearn.linear_model import Lasso
# Creating a pipeline by first standardizing the data and then creating Regression model
pipe_lasso=Pipeline(
[
    ('scalar',StandardScaler()),
    ('reg',Lasso(alpha=0.01))
])
# Training the data by fitting the model on X_train and Y_train 
pipe_lasso.fit(X_train, Y_train)
```

```python
# Creating a variable for Predicted value for Lasso Regression Training set
pred_train_lasso=pipe_lasso.predict(X_train)
# Getting the Metrics for  Lasso Regression Training set
r2_train_lasso,mae_train_lasso,mse_train_lasso,rmse_train_lasso=create_regression_metrics(Y_train,pred_train_lasso)
```

```python
# Creating a variable for Predicted value for Lasso Regression Test set
pred_test_lasso=pipe_lasso.predict(X_test)
# Getting the Metrics for Lasso Regression Test set
r2_test_lasso,mae_test_lasso,mse_test_lasso,rmse_test_lasso=create_regression_metrics(Y_test,pred_test_lasso)
```

```python
# Using Ridge model to train the dataset
from sklearn.linear_model import Ridge
# Creating a pipeline by first standardizing the data and then creating Regression model
pipe_ridge=Pipeline(
[
    ('scalar',StandardScaler()),
    ('reg',Ridge(alpha=0.01))
])
# Training the data by fitting the model on X_train and Y_train 
pipe_ridge.fit(X_train, Y_train)
```

```python
# Creating a variable for Predicted value for Ridge Regression
pred_train_ridge=pipe_ridge.predict(X_train)
# Getting the Metrics for  Ridge Regression Training set
r2_train_ridge,mae_train_ridge,mse_train_ridge,rmse_train_ridge=create_regression_metrics(Y_train,pred_train_ridge)
```

```python
# Creating a variable for Predicted value for Ridge Regression Test set
pred_test_ridge=pipe_ridge.predict(X_test)
# Getting the Metrics for Ridge Regression Test set
r2_test_ridge,mae_test_ridge,mse_test_ridge,rmse_test_ridge=create_regression_metrics(Y_test,pred_test_ridge)
```

#### The results of the Lasso and ridge model indicates that  81% of the variablity of the target variable is explained by the Ridge Regression model, which is similar to Linear regression results. The results of MAE and RMSE are also comparable to simple linear regression model. So all these models appears to be a good fit for the data and  can explain the variablity of the target variables.
